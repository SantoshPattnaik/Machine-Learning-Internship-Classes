## Arithmetic Operations

```python
s1=pd.Series([1,2,3,4,5])
s2=pd.Series([10,20,30,40,50])

print(s1+s2)
print()
print(s1-s2)
print()
print(s1*s2)
print()
print(s2/s1)
print()
print(s2%s1)
print()
print(s1.add(s2))
print()
print(s1.sub(s2))
print()
print(s1.mul(s2))
print()
print(s2.div(s1))
print()
print(s2.mod(s1))
print()
print(s2.pow(s1))
```

```python
student_scores=pd.Series(
    [88,92,75,91,85],
    index=['Bob','Charlie','Alice','David','Eva']
)

student_ranks=student_scores.rank()
print(student_ranks)
print()
student_ranks=student_scores.rank(ascending=False)
print(student_ranks)
print()
student_ranks=student_scores.rank(ascending=False)
print(student_ranks.sort_index())
print()
```

## Checking Null Values

```python
sales_data=pd.Series(
    [250,np.nan,175,300,np.nan,220],
    index=[7,4,3,2,8,9]
)

print(sales_data.isnull())
print()
print(sales_data.notnull())
```

```python
for city, group in temp_df.groupby('City'):
    print(f'City:{city}')
    print(group)
    print()
```

```python
g=temp_df.groupby('City')['Temperature'].mean()
colors=['red','green','blue','yellow']

g.plot(kind='bar',color=colors,title='Mean Temperature by City')
plt.ylabel('Mean Temperature')
plt.xlabel('City')
plt.show()
```
