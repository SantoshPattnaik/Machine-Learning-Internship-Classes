## Range

```python
data_range=iris.max(numeric_only=True)-iris.min(numeric_only=True)
print(f'The range of iris dataset is:\n{data_range}')
```

## Gaussian Distribution

```python
for column in iris.columns:
    plt.figure(figsize=(6,4))
    sns.histplot(iris[column],kde=True,color='blue')
    plt.title(f'Gaussian Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.grid(True)
    plt.show()
```

```python
q1=iris.quantile(0.25,numeric_only=True)
q3=iris.quantile(0.75,numeric_only=True)

iqr=q3-q1
print(f'Interquartile Range:\n{iqr}')
```

## Skewness of attribute distribution
Skewness measure the shape of the distribution.

```python
# calculate skewness
skew_values=iris.skew(numeric_only=True)
print('Skewness of each feature:')
print(skew_values)
```

```python
# plot the distributions to visualize skewness
for column in iris.columns:
    sns.histplot(iris[column],kde=True)
    plt.title(f'Distribution and Skewness of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
