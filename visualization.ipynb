```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_json('iris.json')
obesity=pd.read_csv('Obesity Classification.csv')
obesity.columns
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='species',y='sepalLength',data=df,estimator=np.mean)
plt.title('Mean Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length')
plt.show()
```

```python
plt.figure(figsize=(7,8))
sns.barplot(x='Weight',y='BMI',data=obesity,estimator=np.mean)
plt.title('Average BMI by weight')
plt.xlabel('Weight')
plt.ylabel('Average BMI')
plt.show()
```

```python
# Categorical variable: 'species'
# Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x='species',data=df)
plt.title('Count plot of species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
# count number of each species
species_count=df['species'].value_counts()

# plot pie chart
plt.figure(figsize=(6,6))
plt.pie(species_count,labels=species_count.index,autopct='%1.1f%%',startangle=120)
plt.title('Distribution of Species')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['sepalLength'],bins=10,color='lightgreen',edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
# sns.kdeplot(df['sepalWidth'],color='red')
sns.kdeplot(df['sepalWidth'],fill=True,color='red')
plt.title('Density plot of Sepal Length')
plt.xlabel('Sepal Width')
plt.ylabel('Density')
plt.show()
```

```python
# boxplot
plt.figure(figsize=(6,2))
sns.boxplot(x=df['sepalLength'],color='orange')
# sns.boxplot(y=df['sepalLength'],color='orange')
plt.title('Box plot of sepal length')
plt.xlabel('Sepal Length')
plt.show()
```

```python
sns.scatterplot(x='sepalWidth',y='petalWidth',data=df,hue='species')
plt.title('Scatter Plot: Sepal Width vs Petal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species',loc='center right')
plt.show()
```

```python
# Line plot of Sepal Length over index
sns.lineplot(x=df.index,y='sepalLength',data=df)
plt.title('Line Plot: Sepal Length over Index')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
# box plot
sns.boxplot(x='sepalLength',y='species',data=df)
plt.title('Box plot: Sepal Length by species')
plt.show()
```

```python
# violin plot
sns.violinplot(x='species',y='sepalLength',data=df,hue='species')
plt.title('Violin Plot: Sepal Length by Species')
plt.show()
```

```python
# swarm plot of Sepal Length for each species 
sns.swarmplot(x='species',y='sepalLength',data=df,hue='species')
plt.title('Swarm Plot: Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
# scatter plot
sns.scatterplot(x='sepalLength',y='petalLength',data=df,hue='species')
plt.title('Scatter Plot: Sepal vs Petal Length')
plt.show()
```

## 2D Histogram

```python
plt.hist2d(df['sepalWidth'],df['petalWidth'],bins=30,cmap='Reds')
plt.colorbar(label='count in bin')
plt.title('2D Histogram: Sepal Width vs Petal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
# Line plot of Petal Width for each species over sample index
sns.lineplot(x=df.index,y='petalWidth',hue='species',data=df)
plt.title('Line plot: Petal Width by Species')
plt.xlabel('Sample Index')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# add a new categorical column with 3 flower types
def assign_flower_type(species):
    if species == 'setosa':
        return 'Type A'
    elif species == 'versicolor':
        return 'Type B'
    else:
        return 'Type C'

df['flower_type']=df['species'].apply(assign_flower_type)

# contingency table between species and flower_type
contingency_table=pd.crosstab(df['species'],df['flower_type'])
print(contingency_table)
```

# Multivariate Plot

```python
# compute correlation matrix
corr=df[['sepalLength','sepalWidth','petalLength','petalWidth']].corr()

# plot heatmap
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
